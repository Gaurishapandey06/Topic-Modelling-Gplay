```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')


from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def clean_text(text):
    # lowercase
    text = text.lower()

    # remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)

    # remove emojis
    text = re.sub(r'[^\w\s]', '', text)

    # remove punctuation
    text = re.sub(r'[^\w\s]', '', text)

    # remove numbers
    text = re.sub(r'\d+', '', text)

    # tokenize
    tokens = nltk.word_tokenize(text)

    # remove stopwords + lemmatize
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words and len(word) > 2]

    # join back into clean text
    return " ".join(tokens)

```

```python
df['clean_review'] = df['content'].astype(str).apply(clean_text)
df[['content', 'clean_review']].head(10)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(
    max_df=0.9,   # ignore very common words
    min_df=20     # ignore very rare words
)

dtm = vectorizer.fit_transform(df['clean_review'])

dtm.shape
```

```python
from sklearn.decomposition import LatentDirichletAllocation

lda = LatentDirichletAllocation(
    n_components=7,   # we start with 6 topics
    random_state=42
)

lda.fit(dtm)
```

```python
words = vectorizer.get_feature_names_out()

def show_topics(model, words, n_top_words=10):
    for topic_idx, topic in enumerate(model.components_):
        print(f"\nTopic {topic_idx + 1}:")
        print(", ".join([words[i] for i in topic.argsort()[:-n_top_words - 1:-1]]))

show_topics(lda, words)
```

```python
import numpy as np

topic_probs = lda.transform(dtm)
df['dominant_topic'] = np.argmax(topic_probs, axis=1)

df['dominant_topic'].value_counts()
```

```python
import matplotlib.pyplot as plt

# Count topics
topic_counts = df['dominant_topic'].value_counts().sort_index()

# Human-readable topic names (in the correct order)
topic_names = {
    0: "General Frustration",
    1: "Positive Feedback & Praise",
    2: "Reminder / Task Usage",
    3: "Reminder & Notification Problems",
    4: "Feature Requests & UI",
    5: "Sync & Account Issues",
    6: "Subscription & Pricing Complaints"
}

# Prepare labels
labels = [topic_names[i] for i in topic_counts.index]
values = topic_counts.values

# Plot
plt.figure(figsize=(12,6))
plt.bar(labels, values)
plt.xticks(rotation=45, ha='right')
plt.title("Distribution of User Review Topics (Google Play Productivity Apps)")
plt.ylabel("Number of Reviews")
plt.xlabel("Topic")
plt.tight_layout()
plt.show()
```

```python
# Sort by frequency
topic_counts_sorted = df['dominant_topic'].value_counts()

labels = [topic_names[i] for i in topic_counts_sorted.index]
values = topic_counts_sorted.values

plt.figure(figsize=(12,6))
plt.bar(labels, values)
plt.xticks(rotation=45, ha='right')
plt.title("Most Frequent User Review Topics (Sorted by Importance)")
plt.ylabel("Number of Reviews")
plt.xlabel("Topic")
plt.tight_layout()
plt.show()
```

```python
df['topic_name'] = df['dominant_topic'].map(topic_names)
for topic_num, topic_label in topic_names.items():
    print("\n" + "="*80)
    print(f"TOPIC {topic_num}: {topic_label}")
    print("="*80)

    samples = df[df['dominant_topic'] == topic_num].sample(7, random_state=42)

    for i, review in enumerate(samples['content'], 1):
        print(f"\nExample {i}:")
        print(review[:500])  # print first 500 characters
```

```python
summary = (
    df.groupby('topic_name')
      .agg(
          review_count=('content', 'count'),
          avg_rating=('score', 'mean')
      )
      .sort_values(by='review_count', ascending=False)
)

summary
```
